```python
# Merge credits_df and movies_df on 'title' column
movies_df = movies_df.merge(credits_df, on='title')
```

```python
# Select relevant features (columns)
movies_df = movies_df[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
# Drop rows with missing values
movies_df.dropna(inplace=True)
```

```python
# Text preprocessing functions

```

```python
# Function to convert JSON strings to lists of dictionaries
def convert(obj):
    return [i['name'] for i in ast.literal_eval(obj)]

```

```python
# Function to extract only the first 3 names from the cast list
def convert3(obj):
    L=[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L

```

```python
# Function to fetch director name from crew list
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
# Apply text preprocessing to relevant columns
movies_df['overview'] = movies_df['overview'].apply(lambda x:x.split())
movies_df['genres'] = movies_df['genres'].apply(convert)
movies_df['keywords'] = movies_df['keywords'].apply(convert)
movies_df['cast'] = movies_df['cast'].apply(convert3)
movies_df['crew'] = movies_df['crew'].apply(fetch_director)
```

```python
movies_df['tags'] = movies_df['overview'] + movies_df['genres'] + movies_df['keywords'] + movies_df['cast'] + movies_df['crew']
```
